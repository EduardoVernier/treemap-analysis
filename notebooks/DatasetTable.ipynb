## Create a table with the following columns
Dataset | Revisions | Nodes (total) | Average max depth | Average leaf depth

```python
def generate_row(dataset_id):
    df = Parser.parse_dataset(dataset_id)
    n_revisions = df.shape[1] - 1 # Number of non id columns in data frame

    # Total nodes 
    total_nodes = 0
    for column in df:
        if column != 'id':
            total_nodes += df[column].notnull().sum()

    # Average leaf depth
    avg_leaf_depth = np.mean([item.count('/') + 1 for item in df['id']])

    # Avg max depth
    max_depths = [] 
    for column in df:
        if column != 'id':
            mask = df[column].notnull()
            max_depths.append(np.amax([item.count('/') + 1 for item in df['id'][mask]]))
    avg_max_depth = np.mean(max_depths)
    
    row = pd.DataFrame({'Dataset' : [dataset_id],
                        'Revisions' : [n_revisions],
                        'Nodes (Total)' : [total_nodes],
                        'Avg. leaf depth' : [avg_leaf_depth],
                        'Avg. tree depth' : [avg_max_depth]})
    row = row[['Dataset', 'Revisions', 'Nodes (Total)', 'Avg. leaf depth', 'Avg. tree depth']]
    return row

```

```python
generate_row('exo')
```

```python
rows = []
for dataset_id in Parser.list_datasets():
    rows.append(generate_row(dataset_id))
```

```python
table = pd.concat(rows)
table
```
